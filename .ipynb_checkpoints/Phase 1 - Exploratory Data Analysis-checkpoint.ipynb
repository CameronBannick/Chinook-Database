# Exploratory data analysis

## What countries use the service?

```python
#save query as variable
q = """SELECT Country,
              COUNT(*) AS number_of_customers
       FROM customer
       GROUP BY Country
       ORDER BY number_of_customers DESC;"""
#save query as df
df = pd.read_sql(q, engine)
#call pandas to run query
pd.read_sql(q, engine)
```

```python
#call countplot
sns.barplot(x='Country', y='number_of_customers', data=df)
#set x ticks
plt.xticks(rotation=90)
#set axes labels
plt.xlabel('Country')
plt.ylabel('# of customers')
#set plot title
plt.title('Customers by country')
#show plot
plt.show()
```

The service is currently being used in 24 countries with a number of customers between 1 and 13

## What states use our service in the U.S.?

```python
#save query as variable
q = """SELECT State,
              COUNT(*) AS number_of_customers
       FROM customer 
       WHERE Country = "USA"
       GROUP BY State
       ORDER BY number_of_customers DESC;"""
#save query as df
df = pd.read_sql(q, engine)
#call pandas to run query
pd.read_sql(q, engine)
```

```python
#call countplot
sns.barplot(x='State', y='number_of_customers', data=df)
#set x ticks
plt.xticks(rotation=90)
#set axes labels
plt.xlabel('State')
plt.ylabel('# of customers')
#set plot title
plt.title('Customers by US states')
#show plot
plt.show()
```

The service is currently being used in 11 states where there are 7 users in most states except for the number one, California, has 21.

## What states in Canada use the service

```python
#save query as variable
q = """SELECT State,
              COUNT(*) as number_of_customers
       FROM customer
       WHERE Country = 'Canada'
       GROUP BY State
       ORDER BY number_of_customers DESC;"""
#save query as df
df = pd.read_sql(q, engine)
#call pandas to run query
pd.read_sql(q, engine)
```

```python
#call countplot
sns.barplot(x='State', y='number_of_customers', data=df)
#set x ticks
plt.xticks(rotation=90)
#set axes labels
plt.xlabel('State')
plt.ylabel('# of customers')
#set plot title
plt.title('Customers by Canadian states')
#show plot
plt.show()
```

The service is currently being used in 7 states in Canada with one or two users in each

# What are the summary statistics for the total price of each invoice?

```python
#save query as varaible 
q = """SELECT * 
       FROM invoice;"""
#call pandas to save query as df
df=pd.read_sql(q, engine)
#call head()
df.head()
```

```python
#call describe on total feature
df['Total'].describe()
```

```python
#call boxplot
sns.boxplot(x='Total', data=df)
#set axes labels
plt.xlabel('Invoice total ($)')
#show plot
plt.show()
```

```python
#save summary statistics as variables 
min = np.min(df['Total'])
q1 = np.percentile(df['Total'], 25)
median = np.median(df['Total'])
q3 = np.percentile(df['Total'], 75)
max= np.max(df['Total'])
#call sns hist plot
sns.histplot(data=df['Total'])
#call axvline for summary statistics
plt.axvline(min, color='black', linestyle='solid', label='Min')
plt.axvline(q1, color='blue', linestyle='dashed', label='Q1')
plt.axvline(median, color='black', linestyle='solid', label='Q2 (Median)')
plt.axvline(q3, color='blue', linestyle='dashed', label='Q3')
plt.axvline(max, color='black', linestyle='solid', label='Max')
#call axis title
plt.title('Histogram for invoice total')
#call axis title
plt.xlabel('Total ($)')
#call legend
plt.legend()
#show plot
plt.show()
```

The majority of invoices run between 1 and 8 dollars with the average invoice total of 5.62. However, with the median being only 3.96 it indicates the average is being skewed by some outliers which can be seen in the box plot as black diamonds. The lowest invoices was 0.99 cents and highest invoice was 25.86 dollars.
