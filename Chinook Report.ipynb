# Exploratory data analysis

## What countries use the service?

```python
#save query as variable
q = """SELECT Country,
              COUNT(*) AS number_of_customers
       FROM customer
       GROUP BY Country
       ORDER BY number_of_customers DESC;"""
#save query as df
df = pd.read_sql(q, engine)
#call pandas to run query
pd.read_sql(q, engine)
```

```python
#call countplot
g = sns.barplot(x='Country', y='number_of_customers', data=df)
#set x ticks
plt.xticks(rotation=90)
#set axes labels
plt.xlabel('Country')
plt.ylabel('# of customers')
#set plot title
g.set_title('Customers by country')
#plt.savefig('country')
#show plot
plt.show()
```

The service is currently being used in 24 countries with a number of customers between 1 and 13

```python
#call countplot
g = sns.barplot(x='State', y='number_of_customers', data=df)
#set x ticks
plt.xticks(rotation=90)
#set axes labels
plt.xlabel('State')
plt.ylabel('# of customers')
#set plot title
g.set_title('Customers by US states')
#plt.savefig('customer_usa')
#show plot
plt.show()
```

The service is currently being used in 11 states where there are 7 users in most states except for the number one, California, has 21.

## What type of audio do customers purchase most?

```python
#save query as variable 
q = """SELECT p.Name AS media_type,
              COUNT(*) AS downloads
       FROM invoice AS i 
       LEFT JOIN invoiceline AS il 
       ON i.InvoiceId = il.InvoiceId
       LEFT JOIN track AS t
       ON il.TrackId = t.TrackId
       LEFT JOIN playlisttrack AS pl
       ON t.TrackId = pl.TrackId
       LEFT JOIN playlist AS p
       ON pl.PlaylistId = p.PlaylistId
       GROUP BY media_type
       ORDER BY downloads DESC;"""
#call pandas to run query
df = pd.read_sql(q, engine)
#inspect data 
df.head()
```

```python
#call sns catplot
g = sns.catplot(x='media_type', y='downloads', data=df, kind='bar')
#call title
g.fig.suptitle('Media type downloads')
#rotate xlabels
plt.xticks(rotation=90)
#call axis labels
plt.xlabel('Media genre')
#plt.savefig('media_download')
#show plot
plt.show()
```

The most downloaded media on the site are Music, 90's Music, and TV shows. However, it is overwhelmingly music.

## What are the most downloaded genres?

```python
#save query as Variable 
q = """SELECT g.Name as genre,
              COUNT(*) as downloads
       FROM invoice AS i
       LEFT JOIN invoiceline AS il
       ON i.InvoiceId = il.InvoiceId
       LEFT JOIN track AS t
       ON il.TrackId = t.TrackId
       LEFT JOIN genre AS g
       ON t.GenreId = g.GenreId
       GROUP BY genre
       ORDER BY downloads DESC;"""
#call pandas to run query
df = pd.read_sql(q, engine)
#inspect data 
df.head()
```

```python
#call sns catplot
g = sns.catplot(x='genre', y='downloads', data=df, kind='bar')
#call title
g.fig.suptitle('Genre downloads')
#rotate xlabels
plt.xticks(rotation=90)
#call axis labels
plt.xlabel('Genre')
#plt.savefig('genre_download')
#show plot
plt.show()
```

The top 5 downloaded genres are Rock, Latin, Metal, Alternative/Punk, and Jazz.
