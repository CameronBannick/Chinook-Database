# Exploratory data analysis

```python
#save query as variable
q = """SELECT BillingCountry,
              count(*) AS number_of_customers,
              ROUND(avg(Total), 2) AS average_payments,
              sum(Total) AS total_payments
       FROM invoice
       GROUP BY BillingCountry
       ORDER BY number_of_customers DESC, total_payments DESC;"""
#call pandas to run query
pd.read_sql(q, engine)
```

```python
#save query as variable
q = """SELECT t.Name, count(*) AS downloads, sum(Total) as total_payments
       FROM invoice AS i
       LEFT JOIN invoiceline AS l 
       ON i.InvoiceId=l.InvoiceId
       LEFT JOIN track as t
       on l.TrackId = t.TrackId
       GROUP BY t.Name
       ORDER BY downloads DESC, total_payments DESC
       LIMIT 25;"""
#call pandas to run query
pd.read_sql(q, engine)
```
